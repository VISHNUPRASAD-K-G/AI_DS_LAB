```python
r=int(input("Rows: "))
c=int(input("Col : "))

print("Enter m1")

m1=[[int(input()) for c in range (c)] for r in range(r)]
print('Enter m2:')
m2=[[int(input()) for c in range (c)] for r in range(r)]

print(f"m1=\n{m1}\n\nm2=\n{m2}")
```

```python
import numpy as np

add=np.add(m1,m2)
sub=np.subtract(m1,m2)
div=np.divide(m1,m2)
mult=np.multiply(m1,m2)

print(f"add= {add}\nsub={sub}\ndiv={div}\nmultiply={mult}")
```
