```python
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.metrics import confusion_matrix,precision_score,recall_score,accuracy_score
import matplotlib.pyplot as plt
from sklearn import tree
import pandas as pd
import numpy as np
#data=load_iris()
#x=data.data
#y=data.target

csv=pd.read_csv('diabetes.csv')
x=np.array(csv.iloc[:,:8])
y=np.array(csv.loc[:,'Outcome'])

X_train,X_test,y_train,y_test=train_test_split(x[:100],y[:100],test_size=.20,random_state=89)

dtc=tree.DecisionTreeClassifier()
dtc.fit(X_train,y_train)
pred=dtc.predict(X_test)

print("Accuracy : ",accuracy_score(pred,y_test)*100)

plt.figure(figsize=(10,10))
tree.plot_tree(dtc,filled=True)
plt.show()
```
